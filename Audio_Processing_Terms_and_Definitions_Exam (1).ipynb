## What is a envelope?

In sound and music, an envelope describes how a sound changes over time. It may relate to elements such as amplitude (volume), filters (frequencies) or pitch. For example, a piano key, when struck and held, creates a near-immediate initial sound which gradually decreases in volume to zero.

The most common form of envelope generator is controlled with four parameters: attack, decay, sustain and release (ADSR). The most common kind of envelope generator has four stages: attack, decay, sustain, and release (ADSR)

Attack is the time taken for initial run-up of level from nil to peak, 

1.   The attack is the portion of the envelope that represents the time taken for the amplitude to reach its maximum level. Essentially it is the initial build-up of a sound.
2.   The decay is the progressive reduction in amplitude of a sound over time. The decay phase starts as soon as the attack phase has reached its peak. In the decay phase, the signal level drops until it reaches the sustain level.
3. The sustain is the period of time during which the sound is held before it begins to fade out. Many instruments do not contain a sustain phase.
4. The release is the final fade or reduction in amplitude over time.


While attack, decay, and release refer to time, sustain refers to level.

## **Oscillators**

***Oscillators in Pure Data:***

Oscillators are the basic signal generators in electronic music. By combining, filtering or modulating them, almost any imaginable sound can be created. In Pure Data, audio signals are represented by a stream of numbers between the values of -1 and 1. So the waveform of each oscillator has been programmed to send out values within this range.
